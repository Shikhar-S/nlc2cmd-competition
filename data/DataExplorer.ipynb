```python
from io import StringIO
from html.parser import HTMLParser

import nltk
from nltk import WordNetLemmatizer
from nltk.corpus import stopwords

class MLStripper(HTMLParser):
    def __init__(self):
        super().__init__()
        self.reset()
        self.strict = False
        self.convert_charrefs= True
        self.text = StringIO()
    def handle_data(self, d):
        self.text.write(d)
    def get_data(self):
        return self.text.getvalue()

def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()

def strip_weirdos(unnorm_text):
    tokens = unnorm_text.split()
    return ' '.join(filter(lambda x: x.isalpha(), tokens))

def normalize_text(unnorm_text):
    tokens = unnorm_text.split()
    tokens = map(lambda x: x.lower(), tokens)
    filt_tokens = filter(lambda x: x not in stopwords.words('english'), tokens)
    lemma = WordNetLemmatizer()
    tokens1 = [lemma.lemmatize(word, pos="n") for word in filt_tokens]
    final_tokens = [lemma.lemmatize(word, pos="v") for word in tokens1]
    return ' '.join(final_tokens)
```

```python
for utility in relevant_utilities:
    tp = 0
    fp = 0
    fn = 0
    try:
        for parag in name2man[utility][0]['paragraphs']:
            no_weird_text = strip_tags(parag['text'])
            if no_weird_text.strip()[0] == '-':
                if parag['is_option']:
                    tp += 1
                else:
                    fp += 1
            elif parag['is_option']:
                fn += 1
    except IndexError:
        print(f'Error in utility {utility}')
        continue
    
    precision = tp / (tp + fp) if tp + fp != 0 else 1
    recall = tp / (tp + fn) if tp + fn != 0 else 1
    f1 = 2 / (recall**-1 + precision**-1)
    
    if abs(f1 - 1.) > 0.01:
        print(f'{utility}: {tp} tp, {fp} fp, {fn} fn, total {tp+fp+fn} | p {precision:.1f} r {recall:.1f} f1 {f1:.2f}')
```

### tfidf stuff
it is simpler than you might think

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer()
```

```python
corpus = []
norm_corpus = []
for parag in name2man['find'][0]['paragraphs']:
    if not parag['is_option']: continue
    corpus.append(parag['text'])
    non_weird_text = strip_weirdos(strip_tags(parag['text']))
    norm_text = normalize_text(non_weird_text)
    norm_corpus.append(norm_text)
X = vectorizer.fit_transform(norm_corpus)
```

### Gather nl2bash dataset statistics
What is the average utility number?
How many flags do utilities usually have?

This should give a better idea on what we need to focus on, or at the very least the model is learning through the training data.

```python
from nl2bash.bashlint import data_tools
import tqdm
```

```python
# average utility number
util_num = []
for idx, data_pair in tqdm.tqdm(data.items()):
    cmd = data_pair['cmd']
    cmd_ast = data_tools.bash_parser(cmd)
    util_num.append(len(data_tools.get_utilities_seq(cmd_ast)))
```

```python
def explorer(node):
    print(node.value, node.kind == 'flag')
    for child in node.children:
        explorer(child)

explorer(cmd_ast)
```

```python
plt.hist(util_num, range=(0.5, 11.5), bins=11, density=True)
# plt.yscale('log')
plt.title('Number of utilities in command')
plt.xlabel('# utilities')
plt.ylabel('frequency')
```

```python
# average flag number
def flag_finder(node):
    unresolved_flags = []
    set_results = []
    
    # gather results from children
    for child in node.children:
        ufs, srs = flag_finder(child)
        set_results += srs
        unresolved_flags += ufs
    
    if node.is_utility():
        set_results.append({node.value: unresolved_flags})
        unresolved_flags = []
    elif node.kind == 'flag':
        unresolved_flags.append(node.value)
    
    return unresolved_flags, set_results

util_num = []
for idx, data_pair in tqdm.tqdm(data.items()):
    cmd = data_pair['cmd']
    cmd_ast = data_tools.bash_parser(cmd)
    _, set_results = flag_finder(cmd_ast)
    util_num += set_results
```

```python
num_flags = [len(list(d.values())[0]) for d in util_num]
print('Minimum number of flags:', min(num_flags))
print('Maximum number of flags:', max(num_flags))
plt.hist(num_flags, range=(0, 10), bins=10)
```

```python
# for specific util
query_util = 'ls'
num_flags = [len(list(d.values())[0]) for d in util_num if query_util in d]
print(f'Minimum number of flags for command {query_util}:', min(num_flags))
print(f'Maximum number of flags for command {query_util}:', max(num_flags))
print(f'Mean number of flags for command {query_util}:', sum(num_flags)/len(num_flags))

qu_flags_in_data = sum([list(d.values())[0] for d in util_num if query_util in d], [])
qu_flags_counter = Counter(qu_flags_in_data)
present_num = 5
print(f'Top {present_num} Common Flags:')
for k, v in sorted(list(qu_flags_counter.items()), key=lambda x: x[1], reverse=True)[:present_num]:
    print(k, v)

plt.hist(num_flags, range=(0, 10), bins=10)
```

### Making quick local-eval data

```python
all_evals = []
prev_dict = dict()
prev_dict['invocation'] = ''
prev_dict['cmd'] = []
for key, val in data.items():
    if prev_dict['invocation'] != val['invocation']:
        all_evals.append(prev_dict)
        prev_dict = dict()
        prev_dict['invocation'] = val['invocation']
        prev_dict['cmd'] = [val['cmd']]
    else:
        prev_dict['cmd'].append(val['cmd'])
all_evals.append(prev_dict)
all_evals.pop(0)
    
```

```python
all_evals_dict = dict()
for idx, val in enumerate(all_evals):
    all_evals_dict[str(idx)] = val
```

```python
with open('all-data-eval.json', 'w') as f:
    json.dump(all_evals_dict, f, indent=4)
```
