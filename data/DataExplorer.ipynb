```python
from io import StringIO
from html.parser import HTMLParser

class MLStripper(HTMLParser):
    def __init__(self):
        super().__init__()
        self.reset()
        self.strict = False
        self.convert_charrefs= True
        self.text = StringIO()
    def handle_data(self, d):
        self.text.write(d)
    def get_data(self):
        return self.text.getvalue()

def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()

def strip_weirdos(unnorm_text):
    tokens = unnorm_text.split()
    return ' '.join(filter(lambda x: x.isalpha(), tokens))
```

```python
for parag in name2man['find'][0]['paragraphs']:
    print(strip_weirdos(strip_tags(parag['text'])))
    print('-----')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer()
```

```python
corpus = []
for parag in name2man['find'][0]['paragraphs']:
    non_weird_text = strip_weirdos(strip_tags(parag['text']))
    corpus.append(non_weird_text)
X = vectorizer.fit_transform(corpus)
```

```python
print('Example words:', vectorizer.get_feature_names()[:5])
print('Total number of words:', len(vectorizer.get_feature_names()))
```

### Looking at nl2bash data
Also testing the `bashlex` library

```python
with open('nl2bash-data.json', 'r') as f:
    data = json.load(f)
```

```python
data['2']
```

```python
import bashlex
```

```python
parts = bashlex.parse('cat /boot/config-`uname -r` | grep IP_MROUTE')
for ast in parts:
    print(ast.dump())
```

```python
my_filter = lambda v: v['cmd'].split()[0] == 'find'
query_matches = [v for k, v in data.items() if my_filter(v)]
query_corpus = [v['invocation'] for v in query_matches]



for v in query_matches:
    if '-exec' not in v['cmd']: continue
    if '|' not in v['cmd']: continue
    print('NLC': v['invocation'])
    print(v['cmd'])
```

```python
import numpy as np
```

```python
q_vectors = vectorizer.transform(query_corpus)
q_vector = q_vectors[0]

max_similar_index = np.argmax(cosine_similarity(X, q_vector)[10:])
print('Similar paragraph:')
print(corpus[max_similar_index])
```
